# User Profiling and Segmentation

## Import Data and Initial Exploration



```python
import pandas as pd

# Load the dataset
data = pd.read_csv('user_profiles_for_ads.csv')

# Display basic information
print(data.info())
print(data.describe())
print(data.isnull().sum())
```

## Visualize Demographic Distributions



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the matplotlib figure
plt.figure(figsize=(15, 10))

# Age Distribution
plt.subplot(2, 2, 1)
sns.countplot(data=data, x='Age', palette='Set2')
plt.title('Age Distribution')

# Gender Distribution
plt.subplot(2, 2, 2)
sns.countplot(data=data, x='Gender', palette='Set2')
plt.title('Gender Distribution')

# Education Level Distribution
plt.subplot(2, 2, 3)
sns.countplot(data=data, x='Education Level', palette='Set2')
plt.title('Education Level Distribution')

# Income Level Distribution
plt.subplot(2, 2, 4)
sns.countplot(data=data, x='Income Level', palette='Set2')
plt.title('Income Level Distribution')

plt.tight_layout()
plt.show()
```

## Analyze Device Usage and Online Behavior



```python
# Set up the matplotlib figure
plt.figure(figsize=(15, 10))

# Device Usage
plt.subplot(2, 2, 1)
sns.countplot(data=data, x='Device Usage', palette='Set2')
plt.title('Device Usage Distribution')
plt.tight_layout()
plt.show()
```

## Segment Users with Clustering



```python
# Set up the matplotlib figure
plt.figure(figsize=(15, 10))

# Average time spent online on weekdays vs weekends
data['Total Time Spent Online'] = data['Time Spent Online (hrs/weekday)'] + data['Time Spent Online (hrs/weekend)']
plt.subplot(2, 2, 1)
sns.histplot(data=data, x='Total Time Spent Online', kde=True, color='purple')
plt.title('Total Time Spent Online')

# Likes and Reactions
plt.subplot(2, 2, 2)
sns.histplot(data=data, x='Likes and Reactions', kde=True, color='orange')
plt.title('Likes and Reactions Distribution')

# Click-Through Rates (CTR)
plt.subplot(2, 2, 3)
sns.histplot(data=data, x='Click-Through Rates (CTR)', kde=True, color='red')
plt.title('Click-Through Rates (CTR)')

# Conversion Rates
plt.subplot(2, 2, 4)
sns.histplot(data=data, x='Conversion Rates', kde=True, color='cyan')
plt.title('Conversion Rates')

plt.tight_layout()
plt.show()
```

## Identifying the most common interests among users

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data
data = {
    'Interest': ['Digital Marketing', 'Investing and Finance', 'Music Production', 'Fitness and Wellness', 'Eco-Friendly Living', 'Fashion Modelling', 'Gourmet Cooking', 'Reading and Literature', 'Gaming', 'Pet Care'],
    'Frequency': [180, 165, 160, 155, 150, 145, 140, 135, 130, 155]
}

df = pd.DataFrame(data)

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Frequency', y='Interest', data=df, palette="coolwarm_r")
plt.title('Top 10 User Interests')
plt.xlabel('Frequency')
plt.ylabel('Interest')
plt.show()
```
